For phone number 

```python
df['Phone_Number']=df['Phone_Number'].str.replace('[^a-zA-Z0-9]','', regex=True)
df
```

```python
# df['Phone_Number'].apply(lambda x: x[0:3]+'-'+x[3:6]+'-'+x[6:])
df['Phone_Number']=df['Phone_Number'].astype('str').apply(lambda x: x[0:3]+'-'+x[3:6]+'-'+x[6:])

df
```

```python
df['Phone_Number']=df['Phone_Number'].str.replace('nan--|Na--', '', regex=True)
df
```
